# Data visualization 

```python
df['release_year'].hist(bins=20)
plt.xlabel('Movie relased year')
plt.ylabel('Frequency')
plt.title('Frequency of relased movies over year')
plt.show()
```

```python
df['maturity_rating'].hist()
plt.xlabel('Maturity_rating')
plt.ylabel('Frequency')
plt.title('Frequency of relased maturity_rating')
plt.show()
```

```python
x_axis = df['release_year'] 
y_axis = df['maturity_rating']

plt.bar(df['release_year'], df['maturity_rating'])
plt.title('release_year vs maturity_rating')
plt.xlabel('release_year')
plt.ylabel('maturity_rating')
plt.show()
```

```python
# Replace non-numeric characters and convert to float
df['duration'] = df['duration'].str.replace('[^\\d.]', '', regex=True).astype(float)
```

```python
x_axis = df['duration']
y_axis = df['maturity_rating']
#plt.figure(figsize=(10, 16))
plt.bar(df['maturity_rating'],df['duration'])
plt.title('maturity_rating vs duration')
plt.xlabel('maturity_rating ')
plt.ylabel('duration')
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.bar(df['duration'],df['maturity_rating'])
plt.title('duration vs maturity_rating')
plt.ylabel('maturity_rating ')
plt.xlabel('duration')
plt.show()
```
